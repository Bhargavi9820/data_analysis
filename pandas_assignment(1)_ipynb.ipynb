```python
#1) How many unique  mobile brands are there in the data provided
data=data["phone_name"].unique()
```

```python
len(data)
```

```python
#2) What is the average ram size of all the collected data
data["ram(GB)"].mean()
```

```python
#3Which phone is having the maximum battery capacity
a=data["battery"].max()
data.loc[data["battery"]==a,["phone_name"]]

```

```python
#3>b>which phone is having the minimum battery?
a=data["battery"].min()
data.loc[data["battery"]==a,["phone_name"]]
```

```python
#4>What is the maximum battery capacity provided by Huawei Company
a=data.loc[data["brand"]=="Huawei",["battery"]].max()
a
```

```python
#5>Which Phone brand is providing the highest and lowest ram
a=data["ram(GB)"].min()
b=data.loc[data["ram(GB)"]==a,["phone_name","brand","ram(GB)"]]
c=data["ram(GB)"].max()
d=data.loc[data["ram(GB)"]==c,["phone_name","brand","ram(GB)"]]
print("PHONE BRANDS WITH LOWEST RAM\n",b)
print("PHONE BRANDS WITH HIGHEST RAM\n",d)

```

```python
#6>How many phones has android 7.0 as there operating system
print(len(data.loc[data["os"]=="Android 7.0",["phone_name","os"]]))
data.loc[data["os"]=="Android 7.0",["phone_name","os"]]
```

```python
#7>which is the oldest phone announced and also give the announced dates.
d=data["announcement_date"].min()
data.loc[data["announcement_date"]==d,["brand","announcement_date"]]
```

```python
#8>Which brand produced maximum number of phone in the given data
g=data.groupby("brand")
a=g[["phone_name","brand"]].count()
a.loc["Xiaomi"]
```

```python
#8>b>which brand is having least no.of phones?
g=data.groupby("brand")
a=g[["phone_name","brand"]].count()
a.loc["Google"]
```

```python
#9>how many phones supports 4k video?
print(len(data.loc[data["video_4K"]==True,["brand","phone_name"]]))
data.loc[data["video_4K"]==True,["brand","phone_name"]]
```

```python
#9)b>how many phones supports 8k video?
print(len(data.loc[data["video_8K"]==True,["brand","phone_name"]]))
data.loc[data["video_8K"]==True,["brand","phone_name"]]
```

```python
#10>How many phones supports Li-Po and Li-ion Battery
a=data.loc[data["battery_type"]=="Li-Po",["phone_name"]].count()
b=data.loc[data["battery_type"]=="Li-Ion",["phone_name"]].count()
print("phones that supports Li-po battery\n",a)
print("phones that supports Li-ion battery\n",b)
```

```python
#11>How many sony phones supports 720x1280 resolution
print(len(data.loc[data["resolution"]=="720x1280",["brand","phone_name"]]))
data.loc[data["resolution"]=="720x1280",["brand","phone_name"]]
```

```python
#12>Print a pivot table for finding the average weights ,total weights,based on brand and battery type
pd.pivot_table(data,index="brand",columns="battery_type",values=["weight(g)"],aggfunc=["mean"])
```
